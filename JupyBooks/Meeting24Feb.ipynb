# Model using all features

All features are: 
+ **dummy data**: vector of ones
+ **in_pos** – In-degree (positive edges): number of positive incoming connections
+ **in_neg** – In-degree (negative edges): number of negative incoming connections
+ **out_pos** – Out-degree (positive edges): number of positive outgoing connections
+ **out_neg** – Out-degree (negative edges): number of negative outgoing connections
+ **total_degree** – Total degree: sum of all incoming and outgoing connections
+ **strength_in** – Weighted in-degree: sum of weights of incoming edges
+ **strength_out** – Weighted out-degree: sum of weights of outgoing edges
+ **betweenness** – Betweenness centrality: how often a node lies on shortest paths between others
+ **closeness** – Closeness centrality: how close a node is to all others via outgoing paths
+ **pagerank** – PageRank: Google's algorithm for node importance based on incoming links
+ **transitivity** – Local transitivity: probability that a node's neighbors are also connected (clustering coefficient)
+ **in_coreness** – In-coreness: largest k-core a node belongs to based on in-edges
+ **out_coreness** – Out-coreness: largest k-core a node belongs to based on out-edges
